```python
# explid/exlpore problem
```

```python
import math
```

```python
N = len(dataset)
d = 10

ads_selected = []
numbers_of_selections = [0] * d
sums_of_rewards = [0] * d
total_reward = 0
c =3/2
```

```python
for n in range(N):
    ad = 0
    max_upper_bound = 0
    # pętla szuka reklamy z największym UCB
    for i in range(d):
        if (numbers_of_selections[i] > 0):
            average_reward = sums_of_rewards[i] / numbers_of_selections[i]
            delta = math.sqrt(c * math.log(n+1) / numbers_of_selections[i])
            upper_bound = average_reward + delta
        else:
            upper_bound = 1e400
        # aktualizacja maksa
        if upper_bound > max_upper_bound:
            max_upper_bound = upper_bound
            ad = i
        average_rewads = [sums_of_rewards[i] / numbers_of_selections[i] for i in range(d)]
        deltas = [math.sqrt(c * math.log(n+1) / numbers_of_selections[i]) for i in range(d)]
        plt.errorbar(x=np.arange(d),
                     y=average_rewads,
                     yerr=deltas,
                     fmt='o')
        plt.ylim(-1, 1)
        
    ads_selected.append(ad)
    numbers_of_selections[ad] = numbers_of_selections[ad] + 1
    reward = dataset.values[n, ad]
    sums_of_rewards[ad] = sums_of_rewards[ad] + reward
    total_reward = total_reward + reward

# z każda próba przedział dla danje reklamy rośnie
```
